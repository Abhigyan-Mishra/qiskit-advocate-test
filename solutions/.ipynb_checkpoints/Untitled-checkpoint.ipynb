```python
%matplotlib inline
#%config InlineBackend.figure_format = 'svg'
import numpy as np
from qiskit import QuantumCircuit, execute
from qiskit.transpiler import CouplingMap, PassManager, TransformationPass
from qiskit.transpiler.passes import LookaheadSwap
from qiskit.compiler import transpile
from qiskit.converters import dag_to_circuit

import logging
#logging.basicConfig(level='DEBUG')


def callback_func(pass_, dag, time, property_set, count):
    
    if isinstance(pass_, TransformationPass):
        circuit = dag_to_circuit(dag)
        
        print(pass_.name())
        display( circuit.draw(output='mpl') )
```

```python
qc_adv5_1 = QuantumCircuit(5)
qc_adv5_1.h(3)
qc_adv5_1.cx(0,4)
qc_adv5_1.cx(3,0)
qc_adv5_1.cx(2,4)
qc_adv5_1.draw('mpl')
```

```python
cmap_adv5_2 = CouplingMap(couplinglist = [[0, 1], [1, 2], [2, 3], [3, 4]])
ls = LookaheadSwap(coupling_map = cmap_adv5_2)
pass_manager = PassManager(ls)
qc_adv5_2 = pass_manager.run(qc_adv5_1)
#qc_adv5_2.draw('mpl')
```

```python

qc_adv5_trans = transpile(qc_adv5_1, optimization_level = 0, coupling_map = cmap_adv5_2, routing_method = 'lookahead', callback = callback_func)
qc_adv5_trans.draw('mpl')
```
